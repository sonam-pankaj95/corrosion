```python
import matplotlib.pyplot as plt
import skimage
from skimage.feature import greycomatrix, greycoprops
from skimage import data
import skimage.io as io
import cv2
import numpy as np
import pandas as pd
```

```python

image = skimage.data.load('/home/sonam/Pictures/report/new.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
image1 = skimage.data.load('/home/sonam/Pictures/report/corrosion.jpg')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

```python
image3 = skimage.data.load('/home/sonam/Pictures/report/old.jpg')
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)
```

```python
xs = []
x = []
for i in range(0,100):
    g = greycomatrix(image, [1,i], [np.pi/2], symmetric = True, normed = True)
    entropy = greycoprops(g , 'contrast')
    contrast = greycoprops(g, 'energy')
    xs.append(entropy[1])
    x.append(contrast[1])
ff = np.concatenate((xs,x),axis = 1)
f = np.log(ff)
a = [[1]*100]
y = np.array(a)
ys = y.T
```

```python
fl = np.hstack((f,ys))
df = pd.DataFrame(fl)
final_data = df.drop(df.index[0])
final_data
df.to_csv("out.csv", index=False)
```

```python
xs1 = []
x1 = []
for i in range(0,100):
    g1 = greycomatrix(image1, [1,i], [0], symmetric = True, normed = True)
    entropy1 = greycoprops(g1 , 'contrast')
    contrast1 = greycoprops(g1, 'energy')
    xs1.append(entropy1[1])
    x1.append(contrast1[1])
ff1 = np.concatenate((xs1,x1),axis = 1)
f1 = np.log(ff1)
b = [[0]*100]
y1 = np.array(b)
ys1 = y1.T
```

```python
xs3 = []
x3 = []
for i in range(0,100):
    g3 = greycomatrix(image3, [1,i], [0], symmetric = True, normed = True)
    entropy3 = greycoprops(g3 , 'contrast')
    contrast3 = greycoprops(g3, 'energy')
    xs3.append(entropy3[1])
    x3.append(contrast3[1])
ff3 = np.concatenate((xs3,x3),axis = 1)
f3 = np.log(ff3)
b = [[0.5]*100]
y3 = np.array(b)
ys3 = y3.T
```

```python
fl1 = np.hstack((f1,ys1))
df = pd.DataFrame(fl1)
final_data1 = df.drop(df.index[0])
final_data1
```

```python
fl3 = np.hstack((f3,ys3))
df = pd.DataFrame(fl3)
final_data3 = df.drop(df.index[0])
final_data3
```

```python
X = data[:, 0:2]
Y = data[:, 2]
print (np.max(Y))
print (np.min(Y))
```

```python
import warnings; warnings.simplefilter('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(20,10))
#plt.axes().set_aspect('equal', 'datalim')

pos = np.where(Y == 1)
neg = np.where(Y == 0)
#mid = np.where(Y==0.5)
plt.scatter(X[pos, 0], X[pos, 1], marker='.', color='#666699', s=400)
plt.scatter(X[neg, 0], X[neg, 1], marker='.', color='#aa0000', s=400)

plt.show()
```

```python
from sklearn import svm
svc = svm.SVC(gamma=0.003, C=50, kernel='linear')
svc.fit(X, Y)
```

```python
fig = plt.figure(figsize=(20,10))

plt.scatter(X[svc.support_, 0], X[svc.support_, 1], marker='o', color='c', s=500)
plt.scatter(X[pos, 0], X[pos, 1], marker='.', color='#666699', s=400)
plt.scatter(X[neg, 0], X[neg, 1], marker='.', color='#990000', s=400)
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)
ax.set_title('SVM For training Corrosion texture and non corrosion texture')

ax.set_xlabel('Contrast')
ax.set_ylabel('Entropy')
plt.show()
```
